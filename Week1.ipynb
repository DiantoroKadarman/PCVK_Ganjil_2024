```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://img.freepik.com/free-photo/sunset-silhouettes-trees-mountains-generative-ai_169016-29371.jpg"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                              #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)     #resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)      #convert color to RGB
  final_frame = cv.hconcat((image, image_2))          #concatenate image
  cv2_imshow(final_frame)                             #show image
  print('\n')
```

```python

tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi, " x ", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y]= (255, 255, 255)

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

Resolusi yang tinggi memungkinkan gambar memiliki lebih banyak detail, sedangkan resolusi rendah menghasilkan gambar yang kurang tajam dan detail. Semakin banyak pixel (resolusi tinggi), semakin baik kualitas visual gambar, tetapi juga membutuhkan lebih banyak sumber daya untuk penyimpanan dan pemrosesan.

```python
#Tugas

#2
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis vertikal ditengah image
for y in range (tinggi):
  image_3[y, int((lebar)/2)]= (255, 255, 255)

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)



#membuat garis menyilang ditengah image
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Calculate the slope of the diagonal line
slope = tinggi / lebar

# Draw the diagonal cross lines
for x in range(tinggi):
  for y in range(lebar):
    if abs(x - slope * y) < 1 or abs(x - (tinggi - slope * y)) < 1 :
      image_3[x, y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)


```

```python
#3
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range ((lebar-250), (lebar-200)):
  image_3[int((tinggi)/2),y]= (255, 255, 255)

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#4
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menentukan koordinat kotak di tengah gambar
x_start = int(tinggi / 2) - 40   # Titik awal vertikal (untuk bagian atas kotak)
x_end = int(tinggi / 2)      # Titik akhir vertikal (untuk bagian bawah kotak)
y_start = int(lebar / 2) - 40   # Titik awal horizontal (untuk bagian kiri kotak)
y_end = int(lebar / 2)      # Titik akhir horizontal (untuk bagian kanan kotak)

# Membuat kotak berwarna putih di tengah gambar
for x in range(x_start, x_end):    # Loop untuk bagian vertikal (baris)
    for y in range(y_start, y_end):  # Loop untuk bagian horizontal (kolom)
        image_3[x, y] = (255, 255, 255)  # Mengubah pixel menjadi warna putih

# Menggabungkan kedua gambar secara horizontal
final_frame = cv.hconcat((image_2, image_3))

# Menampilkan hasil gambar
cv2_imshow(final_frame)
plt.imshow(image_3)
```
